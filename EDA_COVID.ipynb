### 확진자 추이 보기

```python
# 날짜 컬럼 정리

COVID_korea['SimpleDate'] = pd.to_datetime(COVID_korea['Date']).dt.strftime('%m-%d')
```

#### 최초발병일: 2020-01-20

```python
COVID_korea[COVID_korea['Confirmed'] == 0]
```

```python
# 최초발병일부터 그래프 그리기
StartCOVID_korea = COVID_korea[COVID_korea['Date'] >= '2020-01-20']
```

```python
# 사이즈 조정
import seaborn as sns
sns.set(rc={'figure.figsize':(20,8)})
```

```python
sns.barplot(x = 'SimpleDate', y = 'Confirmed', data = StartCOVID_korea)
```

```python
sns.barplot(x = 'SimpleDate', y = 'Death', data = StartCOVID_korea)
```

```python
sns.barplot(x = 'SimpleDate', y = 'newConfirmed', data = StartCOVID_korea)
```
